```python
# Indexing and Slicing of vectors(1 D array)
# Indexing and Slicing of matrixes(2 D arrays)
# Boolean Indexing
```

```python
import numpy as np
```

```python
arr = np.arange(11, 21)
arr
```

```python
arr[5]
```

```python
arr[1: 5]
```

```python
arr[3:]
```

```python
arrSlice = arr[0::3]
arrSlice
```

```python
arr_Slice = np.arange(1, 31).reshape(6,5)
arr_Slice
```

```python
arr_Slice[3]
```

```python
arr_Slice[2,2]
```

```python
slice = arr_Slice[0:2,1:3] #slicing of a matrix (1st rows then colomns)
slice
```

```python
slice2 = arr_Slice[3:,3:]
slice2
```

```python
arr_Slice[:,3]
```

```python
#Boolen indexing
arr1 = np.arange(10,21)
arr1
```

```python
boolIndex = arr1 % 2 == 0
boolIndex
```

```python
arr1 = arr1[boolIndex]
arr1
```
