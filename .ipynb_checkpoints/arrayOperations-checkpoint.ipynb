```python
# Deep and Shallow Copy
```

```python
a5 = np.arange(1,21)
a5
```

```python
slice = a5[:5]
slice
```

```python
slice = slice*10
slice
```

```python
a5
```

```python
# By slicing an  array it creates the shallow copy and changing in slice didn't effect the main array,
# b = a.copy() also this is shallow copy
# but by copying the complete array into another variable and made changing in that variable also effects the main array 
# (deep copy)
# b = a (deep copy)
```

```python
# Matric Operations
```

```python
a5 = np.array([[1,2], [3,4]])
a6 = np.array([[5,6], [7,8]])
```

```python
a5
```

```python
# Array Stackinga
# You cannot stack the matrix only vectors can be stacked
```

```python
a7 = np.array([1,2,3])
a8 = np.array([4,5,6])
```

```python
# Vertical Stack, Horizontal Stack, Simple Stack
```

```python
np.vstack((a7,a8))
```

```python
np.hstack((a7,a8))
```

```python
np.column_stack((a7,a8))
```

```python
#Spliting Array
```

```python
a9 = np.arange(16).reshape(4,4)
a9
```

```python
np.hsplit(a9, 2)
```

```python
a10 = np.vsplit(a9, 2)
a10
```

```python
for i in a10:
    print(i)
```
